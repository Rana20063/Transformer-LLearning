```
import os
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import Dataset, DataLoader
from google.colab import files

# CONFIGURATION
SEED = 42
torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

N_SAMPLES = 100  # Increased to 1M
P_TOTAL = 1.0
SNR_RANGE = (-5, 30) # SNR range in dB
DATA_PATH = "noma_dataset_1M.csv"

# CAPACITY CALCULATION
def noma_sum_capacity(h1, h2, alpha, p_total=1.0, snr_linear=10.0):
    p1, p2 = alpha * p_total, (1.0 - alpha) * p_total

    # Dynamic SIC ordering
    if h1 > h2:
        h_s, h_w = h1, h2
        p_s, p_w = p1, p2
    else:
        h_s, h_w = h2, h1
        p_s, p_w = p2, p1

    # Weak user SINR (with interference)
    sinr_w = (p_w * h_w * snr_linear) / (p_s * h_w * snr_linear + 1.0)
    rate_w = np.log2(1.0 + sinr_w)

    # Strong user SINR (SIC applied)
    sinr_s = p_s * h_s * snr_linear
    rate_s = np.log2(1.0 + sinr_s)

    return (rate_w + rate_s), rate_w, rate_s

# DATASET GENERATION
if not os.path.exists(DATA_PATH):
    print("Generating 1M samples with random SNR...")
    records = []
    alpha_grid = np.linspace(0.01, 0.99, 99)

    for _ in range(N_SAMPLES):
        h1, h2 = np.random.exponential(1.0, 2)
        # Random SNR for each sample
        current_snr_db = np.random.uniform(SNR_RANGE[0], SNR_RANGE[1])
        snr_lin = 10**(current_snr_db / 10)

        best_sum_rate = -1.0
        alpha_star = 0.0

        for a in alpha_grid:
            sum_cap, _, _ = noma_sum_capacity(h1, h2, a, P_TOTAL, snr_lin)
            if sum_cap > best_sum_rate:
                best_sum_rate, alpha_star = sum_cap, a

        records.append([h1, h2, current_snr_db, alpha_star])

    df = pd.DataFrame(records, columns=["h1", "h2", "snr_db", "alpha_star"])
    df.to_csv(DATA_PATH, index=False)
else:
    print("Loading existing dataset...")
    df = pd.read_csv(DATA_PATH)

# PREPROCESSING
X = df[["h1", "h2", "snr_db"]].values.astype(np.float32) # 3 Inputs
y = df[["alpha_star"]].values.astype(np.float32)

scaler_x, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2)

class NomaDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X)
        self.y = torch.tensor(y)
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]

train_loader = DataLoader(NomaDataset(X_train, y_train), batch_size=1024, shuffle=True)
test_loader = DataLoader(NomaDataset(X_test, y_test), batch_size=2048)

# TRANSFORMER MODEL
class TransformerAlphaRegressor(nn.Module):
    def __init__(self, d_model=64):
        super().__init__()
        self.embed = nn.Linear(1, d_model)
        # Positional embedding for 3 tokens: h1, h2, SNR
        self.pos_embed = nn.Parameter(torch.zeros(1, 3, d_model))

        enc_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=8, batch_first=True)
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=3)

        self.head = nn.Sequential(
            nn.Linear(d_model, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        # x: (batch, 3) -> (batch, 3, 1)
        x = x.unsqueeze(-1)
        x = self.embed(x) + self.pos_embed
        x = self.transformer(x)
        x = x.mean(dim=1)
        return self.head(x)

model = TransformerAlphaRegressor().to(device)
print(f"Model running on: {device}")

# TRAINING
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005)

print("Training started...")
for epoch in range(10): # 1M samples need fewer epochs
    model.train()
    total_loss = 0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        pred = model(xb)
        loss = criterion(pred, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Avg Loss: {total_loss/len(train_loader):.6f}")

# SAVE AND EVALUATE
torch.save(model.state_dict(), "transformer_noma_1M_model.pth")

model.eval()
y_preds, y_true = [], []
with torch.no_grad():
    for xb, yb in test_loader:
        outputs = model(xb.to(device))
        y_preds.extend(outputs.cpu().numpy())
        y_true.extend(yb.numpy())

y_preds_rescaled = scaler_y.inverse_transform(np.array(y_preds))
y_true_rescaled = scaler_y.inverse_transform(np.array(y_true))

plt.figure(figsize=(10, 5))
plt.plot(y_true_rescaled[:100], 'b-', label='True Optimum Alpha', alpha=0.6)
plt.plot(y_preds_rescaled[:100], 'r--', label='Model Prediction', alpha=0.8)
plt.title('Power Allocation Performance (1M Samples + Variable SNR)')
plt.legend()
plt.show()

mae = np.mean(np.abs(y_preds_rescaled - y_true_rescaled))
print(f"Mean Absolute Error: {mae:.4f}")
print("Max |err| =", float(np.max(np.abs(y_preds_rescaled - y_true_rescaled))))
X_test_orig = scaler_x.inverse_transform(X_test)

n_show = 10  # kaÃ§ Ã¶rnek yazdÄ±rÄ±lsÄ±n
print("\n--- Samples: Theoretical alpha* (brute-force) vs model prediction ---")
for i in range(n_show):
    h1, h2, snr_db = X_test_orig[i]
    alpha_star_theory = float(y_true_rescaled[i, 0])   # teorik/etiket (alpha_star)
    alpha_pred_model  = float(y_preds_rescaled[i, 0])  # model tahmini

    abs_err = abs(alpha_pred_model - alpha_star_theory)
    print(f"[{i:02d}] h1={h1:.4f}, h2={h2:.4f}, snr_db={snr_db:.2f} | "
          f"alpha*={alpha_star_theory:.4f} | alpha_hat={alpha_pred_model:.4f} | |err|={abs_err:.4f}")



```

```
# ============================================================
# Transformer tabanlÄ± basit fonksiyon tahmini (TAM KOD)
# f(x,y,z,h) = x^3 - 2y^3 + 5z + h
# Girdi:  (x, y, z, h)
# ÃÄ±ktÄ±:  f
#
# NOT: Mimari ve eÄitim akÄ±ÅÄ± senin NOMA kodundaki Transformer ile BÄ°REBÄ°R aynÄ±:
# - Linear(1 -> d_model) embedding
# - Positional embedding (1,4,d_model)
# - TransformerEncoder (3 layer, 8 head)
# - Mean pooling
# - Head: Linear -> ReLU -> Linear -> Sigmoid
# - MSELoss + Adam
# - MinMaxScaler (X ve y)
#
# Ä°STEKLERÄ°N:
# - Grafiklerde 100 yerine 20 Ã¶rnek gÃ¶ster
# - FarkÄ± gÃ¶rmek iÃ§in zoom + ayrÄ±ca hata grafiÄi Ã§iz
# ============================================================

import os
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import Dataset, DataLoader
from google.colab import files

# ============================================================
# 1) KONFÄ°GÃRASYON (GPU DESTEKLÄ°)
# ============================================================
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

N_SAMPLES = 300_000                 # istersen 1_000_000 yapabilirsin
DATA_PATH = "toy_function_dataset.csv"

EPOCHS = 15
BATCH_TRAIN = 2048
BATCH_TEST  = 4096

N_SHOW = 20  # grafikte 20 Ã¶rnek gÃ¶ster

# ============================================================
# 2) HEDEF FONKSÄ°YON
# ============================================================
def f_xyz_h(x, y, z, h):
    return x**3 - 2*(y**3) + 5*z + h

# ============================================================
# 3) VERÄ° SETÄ° ÃRETÄ°MÄ° / YÃKLEME
# ============================================================
if not os.path.exists(DATA_PATH):
    print(f"-> Veri seti Ã¼retiliyor... (Cihaz: {device})")

    # AralÄ±klar: kÃ¼p bÃ¼yÃ¼mesini Ã§ok abartmamak iÃ§in makul seÃ§tik
    # x,y: [-2, 2]
    # z,h: [-5, 5]
    x = np.random.uniform(-2, 2, N_SAMPLES)
    y = np.random.uniform(-2, 2, N_SAMPLES)
    z = np.random.uniform(-5, 5, N_SAMPLES)
    h = np.random.uniform(-5, 5, N_SAMPLES)

    f = f_xyz_h(x, y, z, h)

    df = pd.DataFrame({"x": x, "y": y, "z": z, "h": h, "f": f})
    df.to_csv(DATA_PATH, index=False)
    print("-> Kaydedildi:", DATA_PATH)
else:
    print("-> Mevcut veri seti yÃ¼kleniyor...")
    df = pd.read_csv(DATA_PATH)

# ============================================================
# 4) VERÄ° HAZIRLAMA (MinMaxScaler + train/test)
# ============================================================
X = df[["x", "y", "z", "h"]].values.astype(np.float32)
y = df[["f"]].values.astype(np.float32)

scaler_x, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=SEED
)

class ToyDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]

train_loader = DataLoader(ToyDataset(X_train, y_train), batch_size=BATCH_TRAIN, shuffle=True)
test_loader  = DataLoader(ToyDataset(X_test, y_test), batch_size=BATCH_TEST, shuffle=False)

# ============================================================
# 5) TRANSFORMER MÄ°MARÄ°SÄ° (BÄ°REBÄ°R AYNI)
# ============================================================
class TransformerRegressor(nn.Module):
    def __init__(self, d_model=64):
        super().__init__()
        self.embed = nn.Linear(1, d_model)
        self.pos_embed = nn.Parameter(torch.zeros(1, 4, d_model))  # 4 feature: x,y,z,h

        enc_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=8, batch_first=True, dropout=0.1
        )
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=3)

        self.head = nn.Sequential(
            nn.Linear(d_model, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()   # y_scaled 0-1 olduÄu iÃ§in aynÄ± bÄ±rakÄ±yoruz
        )

    def forward(self, x):
        # x: (B, 4)
        x = x.unsqueeze(-1)                 # (B, 4, 1)
        x = self.embed(x) + self.pos_embed  # (B, 4, d_model)
        x = self.transformer(x)             # (B, 4, d_model)
        x = x.mean(dim=1)                   # (B, d_model)
        return self.head(x)                 # (B, 1) in [0,1]

model = TransformerRegressor().to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ============================================================
# 6) EÄÄ°TÄ°M DÃNGÃSÃ
# ============================================================
print(f"-> EÄitim CihazÄ±: {device} | BaÅlÄ±yor...")
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0.0

    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)

        pred = model(xb)
        loss = criterion(pred, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1:02d} | KayÄ±p: {total_loss/len(train_loader):.7f}")

# ============================================================
# 7) TEST + GÃRSELLEÅTÄ°RME (20 Ã¶rnek + ZOOM + HATA GRAFÄ°ÄÄ°)
# ============================================================
model.eval()
y_preds = []
with torch.no_grad():
    for xb, _ in test_loader:
        y_preds.extend(model(xb.to(device)).cpu().numpy())

y_preds = np.array(y_preds).reshape(-1, 1)

# ÃlÃ§ek geri al (gerÃ§ek f Ã¶lÃ§eÄine dÃ¶n)
y_pred_rescaled = scaler_y.inverse_transform(y_preds)
y_true_rescaled = scaler_y.inverse_transform(y_test)

# Ä°lk 20 Ã¶rnek
true_ = y_true_rescaled[:N_SHOW].flatten()
pred_ = y_pred_rescaled[:N_SHOW].flatten()
err_  = np.abs(true_ - pred_)

# Zoom iÃ§in y-limit: min/max etrafÄ±na kÃ¼Ã§Ã¼k pay
ymin = min(true_.min(), pred_.min())
ymax = max(true_.max(), pred_.max())
pad = (ymax - ymin) * 0.10
if pad == 0:
    pad = 1e-6

# Grafik 1: GerÃ§ek vs Tahmin (zoomlu)
plt.figure(figsize=(20, 20))
plt.plot(true_, 'b-o', label='GerÃ§ek f', alpha=0.7)
plt.plot(pred_, 'r--s', label='Transformer Tahmini', alpha=0.9)
plt.title(f'GerÃ§ek vs Tahmin (Ä°lk {N_SHOW} Ã¶rnek) - Zoomlu')
plt.ylabel('f deÄeri')
plt.ylim(ymin - pad, ymax + pad)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()



# ============================================================
# 8) MODEL KAYDET + EXCEL RAPOR (20 Ã¶rnek)
# ============================================================
model_name = "transformer_toy_function.pth"
torch.save(model.state_dict(), model_name)
files.download(model_name)

# Ä°lk 20 test Ã¶rneÄinin ham (Ã¶lÃ§eklenmemiÅ) x,y,z,h deÄerleri
X_test_rescaled = scaler_x.inverse_transform(X_test)

df_detail = pd.DataFrame({
    "Ornek_No": np.arange(N_SHOW),
    "x": X_test_rescaled[:N_SHOW, 0],
    "y": X_test_rescaled[:N_SHOW, 1],
    "z": X_test_rescaled[:N_SHOW, 2],
    "h": X_test_rescaled[:N_SHOW, 3],
    "Gercek_f": true_,
    "Tahmin_f": pred_,
    "Mutlak_Hata": err_,
    "Isaretli_Hata": signed_err
})

excel_name = "toy_function_detayli_analiz_20.xlsx"
df_detail.to_excel(excel_name, index=False)
files.download(excel_name)

print("-> Bitti! Model ve Excel indirildi.")

```
